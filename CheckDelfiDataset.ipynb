## List non-Valid stops (location_type)

```python
nonvalid_stop_times_stops = stops[(stops['stop_id'].isin(stop_times['stop_id'].unique())) & ~(stops['location_type'].isin([0]))]#['stop_id']
nonvalid_stop_times_stops
```

```python
nonvalid_stop_times_stops_nonan = nonvalid_stop_times_stops[nonvalid_stop_times_stops['parent_station'].notna()]
nonvalid_stop_times_stops_nonan
```

### select parent_station-ID for stop_id (only if parent_station ist of type 0)

```python
filtered_stops = nonvalid_stop_times_stops.reset_index().set_index('parent_station').join(
    stops.loc[stops['stop_id'].isin(nonvalid_stop_times_stops['parent_station'])].reset_index().set_index('stop_id')[['stop_name','location_type']],
    rsuffix='_stops'
).rename_axis('parent_station').reset_index().set_index('index')
```

```python
filtered_stops = filtered_stops[filtered_stops['parent_station'].notna()]
filtered_stops
```

### Alter the original dataset (with parent_stop-IDs instead of the faulty ones) - using .loc

```python
stop_times.loc[stop_times['stop_id'].isin(filtered_stops['stop_id']), 'stop_id'] = stop_times.loc[stop_times['stop_id'].isin(filtered_stops['stop_id']), 'stop_id'].apply(lambda x: filtered_stops.set_index('stop_id')['parent_station'].loc[x])
```

### Delete all stops in stop_times that have no stop_id

```python
stop_times.drop(labels=stop_times[stop_times['stop_id'].isna()].index, inplace=True)
```

## Find non existent stops in transfers.txt

```python
broken_transfers = pd.concat(
    [
        transfers[~transfers['from_stop_id'].isin(stops['stop_id'].unique())],
        transfers[~transfers['to_stop_id'].isin(stops['stop_id'].unique())]
    ]
)
broken_transfers
```

## Exporting Dataset back to csv

```python
import csv
level = csv.QUOTE_ALL #this is actually wrong, but the closest pandas/csv-module can do. Correct would be to quote floats but don't quote integers...very odd conformity...
level
```

```python
import zipfile
import os

iteration='1'

zipfilepath = os.path.join(_base,r"20220509_fahrplaene_gesamtdeutschland_gtfs_fix"+iteration+".zip") #<-- Fix this, if working on a new dataset
zipfilepath
```

```python
zip_ = zipfile.ZipFile(zipfilepath,'w',compression=zipfile.ZIP_DEFLATED)
```

```python
copy_ = [
    os.path.join(_base, "shapes.txt"),
    os.path.join(_base, "agency.txt"),
    os.path.join(_base, "Beschreibung_DELFI-Datensatz_GTFS_20220509.pdf"), #<- FIX this, if you work on a new dataset
    os.path.join(_base, "calendar.txt"),
    os.path.join(_base, "calendar_dates.txt"),
    os.path.join(_base, "frequencies.txt"),
    os.path.join(_base, "levels.txt"),
    os.path.join(_base, "pathways.txt"),
    os.path.join(_base, "routes.txt")
]
for c in copy_:
    zip_.write(
        c,
        os.path.basename(c)
    )
```

```python
copy_ = [
    os.path.join(_base, "trips.txt")
]
for c in copy_:
    zip_.write(
        c,
        os.path.basename(c)
    )
```

```python
stops.to_csv(os.path.join(_base, "stops_new.txt"), index=False, quoting=level )
zip_.write(
    os.path.join(os.path.dirname(zipfilepath),"stops_new.txt"),
    'stops.txt'
)
```

```python
transfers.to_csv(os.path.join(_base,"transfers_new.txt"), index=False, quoting=level )
zip_.write(
    os.path.join(os.path.dirname(zipfilepath),"transfers_new.txt"),
    'transfers.txt'
)
```

```python
stop_times.to_csv(os.path.join(_base, "stop_times_new.txt"), index=False, quoting=level )
zip_.write(
    os.path.join(os.path.dirname(zipfilepath),"stop_times_new.txt"),
    'stop_times.txt'
)
```
